```python
def load_results_directory(results_directory):
    """
    Load all the data files from a given results directory,
    into a dictionary, which is returned.
    """
    csvs = [c for c in os.listdir(results_directory) if c.endswith('.csv')]
    print(f"-- {results_directory} --")
    print('\n'.join(sorted(csvs)))
    datas = {}
    for csv_filename in csvs:
        try:
            datas[csv_filename[:-4]] = pd.read_csv(os.path.join(results_directory, csv_filename))
        except:
            print(f"Couldn't read from {csv_filename}")
            continue
    return datas
    datas.keys()
```

```python
cat_area_per_vol_options = list(map(int,[3e3, 3e4, 3e5, 3e6, 3e7, 3e8, 3e9])) # m2/m3
cat_area_per_vol_options
```

```python
datas = {}
for c in cat_area_per_vol_options:
    directory = f"PFR_results/A{c:.0f}"
    datas[int(c)] = load_results_directory(directory)
datas.keys()
```

```python

def xlabels():
    """
    Probably not used, any more, if you can make the plots
    use the "Distance (mm)" as the x axis.
    """
    ticks = []
    labels = []
    plt.xlabel("Distance down reactor (mm)")
    
    if plt.gca().xaxis.get_scale() == 'log':
        mm = length_mm / number_of_reactors
        while mm <= length_mm:
            ticks.append( (mm * number_of_reactors/length_mm ) )
            labels.append( f'{mm:.1g}')
            mm *= 10
        plt.xticks(ticks, labels)
    else:
        mm = 0
        while mm <= length_mm:
            ticks.append( int(mm * number_of_reactors/length_mm ) )
            labels.append( str(mm) )
            mm += 1
        #labels[-1] = labels[-1] + ' mm'
        plt.xticks(ticks, labels)
```

```python
ax1 = data['T (C)'].plot(logx=True)
plt.ylabel('Temperature (C)')
xlabels()
plt.legend()
ax2 = ax1.twinx()
data['alpha'].plot(ax=ax2, color='tab:orange')
ax2.set_ylim(-2, 2)
plt.legend()
plt.ylabel('alpha')
plt.tight_layout()
plt.savefig('temperature-and-alpha.pdf')
plt.show()
```
